```python
### Create a toy data frame to work with
import pandas as pd
```

```python
id = range(1, 11)
value = [3*e + 1 for e in id]
df = pd.DataFrame({'id':id, 'value':value})
```

```python
# by default, the index is printed out
df
```

```python
# this will not print the index
print(df.to_string(index=False))
```

```python
# this will not print the index (but keep the nice jupyter display format)
from IPython.display import display, HTML
display(HTML(df.to_html(index=False)))
```

```python
# old index kept 
df.sample(frac=1)
```

```python
# new index generated (old index kept)
df.sample(frac=1).reset_index()
```

```python
# new index generated (old index removed)
df.sample(frac=1).reset_index(drop=True)
```

```python
# after shuffle only take 5 (this is the same as randomly selecting 5 from the original dataframe)
df_take5 = df.sample(frac=1).reset_index(drop=True)[:5]
df_take5
```
